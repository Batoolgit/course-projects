```python
import numpy as np
from collections.abc import Callable
import mesh_function as mf
import matplotlib.pyplot as plt
```

```python
deltat = 0.1
# create a erray t where the first value is 0 and the last is 4 and the step is deltat
t = np.arange(0, 4 + deltat, deltat)
f1= mf.mesh_function(mf.func, t[0:31])
f2 = mf.mesh_function(mf.func, 3*t[31:])
f= np.concatenate((f1, f2))
f

# plot f to t using matplotlib

plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f(t)')
# in the same plot plot exp(-t) and exp(-3t)
plt.plot(t, np.exp(-t), label='exp(-t)',linestyle='dashed')
plt.plot(t, np.exp(-3*t), label='exp(-3t)',linestyle='dashed')
plt.legend()    

plt.title('Plot of f(t) from task 1.1b)')
plt.grid()
plt.show()
```
